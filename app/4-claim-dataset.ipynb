```python
import json
import glob
from datasets import Dataset

data = []

for file in glob.glob("../data/0-scraped/*.jsonl"):
    with open(file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue  # skip empty lines
            try:
                obj = json.loads(line)
                # Extract
                first_claim = ""
                if isinstance(obj.get("c"), dict):
                    first_claim = obj["c"].get("1") or ""
                cpc = obj.get("cpc") or []
                if not isinstance(cpc, list):
                    cpc = [cpc] if cpc else []
                cpc = [str(code) for code in cpc]
                data.append({"first_claim": first_claim, "cpc": cpc})
            except Exception as e:
                print(f"Skipping bad line in {file}: {e}")

# Create dataset
dataset = Dataset.from_list(data)

# Push to HF
dataset.push_to_hub("mhurhangee/ep-b1-claims-with-cpc")
```
